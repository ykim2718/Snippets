```python
import pandas as pd
a_date = pd.Timestamp('20170710')
di = pd.date_range(start='20170601', end='20170730', freq='5D')
print(type(di))
di
```

```python
if a_date >= di.min():
    j = di.get_loc(a_date, method='ffill')  # find previous index, i.e,, earlier
else:
    j = 0
di[j]
```

```python
if a_date <= di.max():
    j = di.get_loc(a_date, method='bfill')  # find next index, i.e,, later
else:
    j = -1
di[j]
```

```python
import numpy as np
j = np.abs(np.array(di.tolist()) - a_date).argmin()
di[j]
```
